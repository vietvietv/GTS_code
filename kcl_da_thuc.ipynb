```python
import numpy as np

def tim_khoang_cach_li(he_so):
    """
    he_so: Danh sách các hệ số của đa thức từ bậc cao nhất đến bậc 0.
    Ví dụ: f(x) = x^5 - 3x^3 + 2x^2 - x + 5 -> he_so = [1, 0, -3, 2, -1, 5]
    """
    p = np.poly1d(he_so)
    
    # 1. Tìm giới hạn miền chứa nghiệm R
    a_n = abs(he_so[0])
    max_a_k = max(abs(x) for x in he_so[1:])
    R = 1 + max_a_k / a_n
    print(f"[*] Miền chứa nghiệm tuyệt đối: [-{R:.3f}, {R:.3f}]")
    
    # 2. Tìm các điểm cực trị (nghiệm của đạo hàm p' = 0)
    p_dao_ham = p.deriv()
    nghiem_dao_ham = np.roots(p_dao_ham)
    
    # Lọc lấy các điểm cực trị thực và nằm trong đoạn [-R, R]
    cuc_tri_thuc = nghiem_dao_ham[np.isreal(nghiem_dao_ham)].real
    diem_kiem_tra = [x for x in cuc_tri_thuc if -R <= x <= R]
    
    # 3. Tạo tập hợp các điểm mút để xét dấu
    diem_kiem_tra.extend([-R, R])
    diem_kiem_tra = sorted(list(set(diem_kiem_tra)))
    
    # 4. Xét dấu f(x) để tìm khoảng cách li
    khoang_cach_li = []
    print("[*] Kết quả khảo sát các khoảng đơn điệu:")
    
    for i in range(len(diem_kiem_tra) - 1):
        x1, x2 = diem_kiem_tra[i], diem_kiem_tra[i+1]
        f1, f2 = p(x1), p(x2)
        
        # Nếu tích f1 * f2 < 0, đồ thị cắt trục hoành tại 1 điểm duy nhất
        if f1 * f2 < 0:
            khoang_cach_li.append((x1, x2))
            print(f"    -> Phát hiện khoảng cách li: ({x1:.4f}, {x2:.4f}) | f(x1)={f1:.2f}, f(x2)={f2:.2f}")
        elif f1 == 0:
            print(f"    -> Phát hiện nghiệm đúng: x = {x1:.4f}")
            
    # Kiểm tra mút cuối cùng
    if p(diem_kiem_tra[-1]) == 0:
        print(f"    -> Phát hiện nghiệm đúng: x = {diem_kiem_tra[-1]:.4f}")
        
    return khoang_cach_li

# Chạy thử với đa thức Câu 6: x^5 - 3x^3 + 2x^2 - x + 5 = 0
he_so_cau_6 = [1, 0, -3, 2, -1, 5]
he_so_cau_4 = [ 1, 0, -4, 0, 1]
tim_khoang_cach_li(he_so_cau_6)
```
