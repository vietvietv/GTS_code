```python
import numpy as np

# 1. Khởi tạo mảng dữ liệu X_t (17 điểm)
X = np.array([10, 12, 11, 15, 18, 20, 22, 25, 23, 26, 28, 30, 29, 31, 35, 34, 38])

# 2. Khởi tạo vector trọng số a_j (Spencer 15)
# Ghi chú: Array này tương ứng từ a_(-7) đến a_7
a = np.array([-3, -6, -5, 3, 21, 46, 67, 74, 67, 46, 21, 3, -5, -6, -3]) / 320

# 3. Tính m_hat bằng phép chập (Convolution)
# mode='valid' chỉ tính toán khi cửa sổ quét đủ 15 phần tử dữ liệu
m_hat = np.convolve(X, a, mode='valid')

# 4. Tính phần dư Y_hat
# Do m_hat chỉ tính được cho t=8, t=9, t=10, ta trích xuất X_t tương ứng
Y_hat = X[7:10] - m_hat
```
